```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textblob import TextBlob

# Membaca dataset
# Ganti 'your_dataset.csv' dengan path file dataset Anda
try:
    df = pd.read_excel('clean_filmindo23.xlsx')
except UnicodeDecodeError:
    # Coba baca dengan encoding berbeda jika terjadi error
    df = pd.read_excel('clean_filmindo23.xlsx', encoding='latin1')

# Pastikan kolom 'full_text' ada di dataframe
if 'full_text' not in df.columns:
    raise ValueError("Kolom 'full_text' tidak ditemukan di dataset")
# df = pd.read_csv('clean_filmindo23.xlsx')

# Fungsi untuk melakukan analisis sentimen
def sentiment_analysis(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positif'
    elif analysis.sentiment.polarity == 0:
        return 'Netral'
    else:
        return 'Negatif'

```

```python
# Menerapkan fungsi analisis sentimen pada kolom 'full_text'
df['Sentimen'] = df['full_text'].apply(sentiment_analysis)

# Menghitung jumlah nilai untuk setiap sentimen
sentiment_counts = df['Sentimen'].value_counts()

```

```python
# Membuat bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values)
plt.title('Distribusi Sentimen')
plt.xlabel('Sentimen')
plt.ylabel('Jumlah')
plt.show()
```

```python
# Membuat pie chart
plt.figure(figsize=(8, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proporsi Sentimen')
plt.show()
```

```python
# Menampilkan teks dengan sentimen tertentu
def display_texts(sentiment):
    texts = df[df['Sentimen'] == sentiment]['full_text']
    print(f"Teks dengan Sentimen {sentiment}:")
    for text in texts:
        print(f"- {text}")

# Menampilkan teks untuk masing-masing kategori sentimen
display_texts('Positif')
display_texts('Negatif')
display_texts('Netral')
```

```python
# Menampilkan contoh teks untuk masing-masing sentimen
for sentimen in ['Positif', 'Negatif', 'Netral']:
    print(f"\nContoh teks dengan sentimen {sentimen}:")
    examples = df[df['Sentimen'] == sentimen]['full_text'].head(3)
    for i, text in enumerate(examples, 1):
        print(f"{i}. {text}")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import matplotlib.pyplot as plt

# Mengunduh stopwords dari NLTK
nltk.download('punkt')
nltk.download('stopwords')

# Fungsi untuk menghitung frekuensi kata
def word_frequency(texts):
    words = []
    for text in texts:
        for word in word_tokenize(text):
            words.append(word.lower())

    # Menghilangkan stopwords
    stop_words = set(stopwords.words('english'))  # Sesuaikan dengan bahasa yang relevan
    filtered_words = [word for word in words if word not in stop_words and word.isalpha()]

    return Counter(filtered_words)

# Menghitung frekuensi kata
word_freq = word_frequency(df['full_text'])

# Menampilkan 10 kata teratas
most_common_words = word_freq.most_common(10)
print(most_common_words)

# Membuat plot untuk kata teratas
plt.figure(figsize=(10, 6))
plt.bar(*zip(*most_common_words))
plt.title("Top 10 Most Frequent Words in Tweets")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# Membuat document-term matrix
vectorizer = CountVectorizer(max_df=0.95, min_df=2, stop_words='english')
dtm = vectorizer.fit_transform(df['full_text'])

# Melakukan LDA
lda = LatentDirichletAllocation(n_components=5, random_state=0)
lda.fit(dtm)

# Menampilkan topik-topik
for index, topic in enumerate(lda.components_):
    print(f'Top 10 words for topic #{index}:')
    print([vectorizer.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print('\n')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# Membuat TF-IDF matrix
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf_vectorizer.fit_transform(df['cleaned_text'])

# Melakukan clustering
num_clusters = 5  # Tentukan jumlah cluster sesuai kebutuhan
km = KMeans(n_clusters=num_clusters)
km.fit(tfidf_matrix)

# Menampilkan kata kunci untuk setiap cluster
print("Top terms per cluster:")
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = tfidf_vectorizer.get_feature_names_out()
for i in range(num_clusters):
    top_terms = [terms[ind] for ind in order_centroids[i, :10]]
    print(f"Cluster {i}: {', '.join(top_terms)}")
```
